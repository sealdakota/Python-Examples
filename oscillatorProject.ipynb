<h3>
Simple Harmonic Oscillator
</h3>

```python
#Dakota Seal
import numpy as np
import matplotlib.pyplot as plt

def f(r, t, k):
    theta=r[0]        #unpack the array r into local variables
    omega=r[1]
    #calculated derivatives 
    dtheta = omega
    domega = -k*np.sin(theta)
    #domega = Pendulum equation of motion 
    return np.array([dtheta,domega],float)

start = 0.0
end = 15.0
numSteps = 1500
stepSize = (end-start)/numSteps

g = 9.8 #Gravity
L = int(input("length "))#1.0 #Pendulum Length
k = g/L


tpoints = np.arange(start,end,stepSize)
omegapts = []
thetapts = []

# initial conditions for omega and theta, summarized in vector r
#Theta starts at 1, omega starts at 0.
r = np.array([1.0,0.0],float)   

# all steps are as normal RK4, but vector valued quantities used
for t in tpoints:
    thetapts.append(r[0])
    omegapts.append(r[1])
    #4th order Runge-Kutta method
    k1 = stepSize*f(r,t,k)
    k2 = stepSize*f(r+0.5*k1,t+0.5*stepSize,k)
    k3 = stepSize*f(r+0.5*k2,t+0.5*stepSize,k)
    k4 = stepSize*f(r+k3,t+stepSize,k)
    r += (k1+2.*k2+2.*k3+k4)/6

plt.plot(tpoints,thetapts,label='theta')
plt.plot(tpoints,omegapts,label='omega')
plt.title("Simple Harmonic Oscillator")
plt.xlabel("t")
plt.legend(loc='upper right')
plt.show()
```

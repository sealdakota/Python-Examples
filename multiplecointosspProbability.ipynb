```python
#Dakota Seal
#Show's that statistically it takes less flips to consecutively get 'heads, tails, tails' than to get 'heads, tails, heads'
import matplotlib.pyplot as plt
import numpy as np
import random


samples = ('000', '001', '010', '011', '100', '101', '110', '111')
tailsN = ('0','1','2','3')
x = np.zeros([8])
x2 = np.zeros([4])

for i in range(10000):
    a = random.randint(0,1)
    b = random.randint(0,1)
    c = random.randint(0,1)
    if a == 0:
        if b == 0:
            if c == 0:
                x[0] += 1
                x2[0] += 1
            if c == 1:
                x[1] += 1
                x2[1] += 1
        if b == 1:
            if c == 0:
                x[2] += 1
                x2[1] += 1
            if c == 1:
                x[3] += 1
                x2[2] += 1
    if a == 1:
        if b == 0:
            if c == 0:
                x[4] += 1
                x2[1] += 1
            if c == 1:
                x[5] += 1
                x2[2] += 1
        if b == 1:
            if c == 0:
                x[6] += 1
                x2[2] += 1
            if c == 1:
                x[7] += 1
                x2[3] += 1

fig = plt.figure(figsize=(18,5))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

ax1.bar(samples,x)
ax1.set_xlabel('Coin flip samples')
ax1.set_ylabel('Occurrences')
ax1.set_title('Random coin toss statistics')

ax2.bar(tailsN, x2)
ax2.set_xlabel('outcomes')
ax2.set_ylabel('occurrences')
ax2.set_title('number of tails per set')



def coinToss():
    result = str(random.randint(0,1))
    return result


string1 = ('')
string = ('')
flips = 0
HTTdata = np.array([])
HTHdata = np.array([])
foundHTT = -1
foundHTH = -1

for i in range (1000):
    while foundHTT < 0:
        flips += 1
        if len(string1)<3:
            string1 = string1 + coinToss()
        else:
            string1 = string1[1:]+coinToss()
        if string1 == '011':
            foundHTT = flips
            HTTdata = np.append(HTTdata, foundHTT)
    flips = 0
    foundHTT = -1
    string1 = ('')
    
for i in range (1000):
    while foundHTH < 0:
        flips += 1
        if len(string)<3:
            string = string + coinToss()
        else:
            string = string[1:]+coinToss()
        if string == '010':
            foundHTH = flips
            HTHdata = np.append(HTHdata, foundHTH)
    flips = 0
    foundHTH = -1
    string = ('')

avg1 = np.average(HTTdata)
avg2 = np.average(HTHdata)
print("average HTT flips:")
print(np.average(HTTdata))
print("average HTH flips:")
print(np.average(HTHdata))
print("standard deviation HTT flips:")
print(np.std(HTTdata))
print("standard deviation HTH flips:")
print(np.std(HTHdata))

ax3.hist(HTHdata, alpha = .9)
ax3.hist(HTTdata, alpha = .69)
ax3.set_title("Historgram of average flips, HTH blue, HTT orange")
ax3.set_xlabel("flips")
ax3.set_ylabel("occurences")


plt.show()



```

```python
#Dakota Seal
#Plotting Temperature and humidity vs time of provided weather data and showing avrg, max, and min values

import matplotlib.pyplot as plt
import numpy as np
import csv

time = []
temp = []
humid = []

#opens weather data file and sepperates columns by comma. appends time, temperature, and humidity with data from file. 
with open('weatherData.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        temp.append(float(row[1]))
        time.append(float(row[0]))
        humid.append(float(row[2]))
        
#plots temp vs. time.
plt.plot(time,temp)
plt.ylabel('Temperature (C)')
plt.xlabel('Unix time stamp (s)')
plt.title('temperature vs. time')
plt.show()

#plots humidity vs. time.
plt.plot(time,humid)
plt.ylabel('Humidity (percentage)')
plt.xlabel('Unix time stamp (s)')
plt.title('humidity vs. time')
plt.show()

#Prints average temperature and humidity. Prints min and max temperature. 
print("the average temperature was: " + str(np.sum(temp)/np.size(temp)))
print("the average humidity was: " + str(np.sum(humid)/np.size(humid)))
print("the max temperature was: " + str(np.max(temp)))
print("the min temperature was: " + str(np.min(temp)))
```

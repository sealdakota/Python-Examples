```python
#Dakota Seal
#ploting 8th order interpolating polynomial and showing examples of fit using splines
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

#data points
x = np.array([-4,-3,-2,-1,0,1,2,3,4])
y = np.array([1.125e-7,1.234e-4,1.832e-2,3.679e-1,1,3.679e-1,1.832e-2,1.234e-4,1.125e-7])

#polynomial fit of points with order as x-1
pol = np.polyfit(x,y,len(x)-1)

#value of polynomial using linspace
xline = np.linspace(min(x),max(x))
yline = np.polyval(pol,xline)

#interpolation plots
plt.plot(xline, yline, 'r--',label="interpolation")
plt.plot(x,y,'o',label="data")

#splines
splineCubic = interp1d(x,y,kind = 'cubic')
splineLinear = interp1d(x,y,kind = 'linear')

xRange = np.arange(-4,4,.1)

#spline plots w/ labels
plt.plot(xRange,splineCubic(xRange),'m--',label="cubic spline")
plt.plot(xRange,splineLinear(xRange),'g--',label="linear spline")
plt.legend(loc='lower center', fontsize=16, labelspacing=0.3)
plt.title("8th order interpolating polynomial with splines")


```
